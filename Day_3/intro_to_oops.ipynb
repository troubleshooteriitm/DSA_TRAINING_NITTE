# What is Object-Oriented Programming (OOP)?


- OOP is a programming paradigm that organizes code around "objects" rather than just functions and logic.eg: a `Car` object with properties like `color` and methods like `drive()`.
- It allows us to model real-world entities and their interactions in code, making it easier to manage complexity and promote code reuse.
