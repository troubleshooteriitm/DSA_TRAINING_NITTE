```python
# for n number of inputs
def add_n(*args):
    print(type(args))
    op=0
    for i in args:
        op+=i
    return op

print(add_n(1,2,3,4))
print(add_n(1,2,3,4,5,6))

```

```python
# global keyword
x=10
def fun(y):
    global x
    x+=y
    return x

print(x)
fun(20)
fun(25)
print(x)
```

```python
from functools import reduce


# Sum of numbers
nums = [1, 2, 3, 4, 5]
total = reduce(lambda a, b: a + b, nums)
print(total)
```

```python
# recursion 

def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)# 5*4*3*2*1
```

```python
# factorial using for loop
def factorial(n):
    op=1
    for i in range(1,n+1):
        op*=i
    return op

print(factorial(5))
```

```python
# generator function
def count_up_to(n):
    count=1
    print("before while loop")
    while count<=n:
        yield count
        print(f"Yielded: {count}")
        count+=1
    print("done")

# count_up_to(10)
for num in count_up_to(5):
    print('inside for : "' + str(num) + '"')

# examples of inbuilt generators
# range
# enumerate
# zip
# generators are used in production like in data processing, web scraping, etc. where we need to process large data sets without loading everything into memory at once.
```
