```python
class Person:
    def __init__(self, name, age=20):# constructor
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

p1=Person("Alice", 30)
p2=Person("Bob")
print(p1.greet())  # Output: Hello, my name is Alice and I am 30 years old.
print(p2.greet())  # Output: Hello, my name is Bob and I am 20 years old.
```
